```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from matplotlib.pyplot import errorbar
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("error_wiki_test.csv")
df.head()
```

```python
#y=x
x=range(0,101)

# Plot for Wiki
plt.figure(figsize=(10,10))
plt.plot(x, x, color = 'r')
plt.scatter(df['Actual'], df['Prediction'], color = 'b', alpha=0.2) # Predicted on x_test
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Wiki)')
plt.show()
```

```python
dmean = df.groupby(['Actual'], as_index=False)['Prediction'].mean()
print (dmean.head())
```

```python
# y=x
x=range(0,101)

# Plot average prediction for each actual age
plt.figure(figsize=(10,10))
plt.plot(x, x, color = 'r')
plt.scatter(dmean['Actual'], dmean['Prediction'], color = 'b', alpha=0.2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted (Wiki with mean)')
plt.show()
```
